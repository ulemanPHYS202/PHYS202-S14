```python
def EulerFreefall():
    tmax=10
    dt=0.1
    a=-9.81
    nsteps=int(tmax/dt)
    t=np.zeros(nsteps)
    v=np.zeros(nsteps)        
    y=np.zeros(nsteps)
    t[0]=0.
    v[0]=0.
    for i in range(len(t)-1):
        t[i+1]=t[i]+dt
        v[i+1]=v[i]+dt*a
        y[i+1]=y[i]+dt*v[i]
    return t,v,y
te,ve,ye =EulerFreefall()
```

```python
plt.plot(te,ye,'r-',label='y(t)')
plt.plot(te,ve,'b-',label='v(t)')
plt.xlabel('time (s)',fontsize=20)
plt.ylabel('y(t),v(t)',fontsize=20)

plt.grid()
plt.legend(loc='best',fontsize=20)
plt.show()
```

```python
def freefall(tmax=10.,dt=0.1,t0=0.,v0=46.,y0=0.):
    a = -9.81
    t = np.zeros(int(tmax/dt))
    v = np.zeros(int(tmax/dt))
    y = np.zeros(int(tmax/dt))
    t[0] = t0
    v[0] = v0
    y[0] = y0

    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1] = v[i] + dt * a
        y[i+1] = y[i] + dt * v[i]
    
    tmax = -v0/a
    ymax = y0 + v0*tmax + 0.5*a*tmax**2
    percentDiff = 100.*np.abs(y.max()-ymax)/ymax
    print "ymax percentDiff = %.2f"%percentDiff

    plt.plot(t,y,"r-",label="y(t)")
    plt.plot(t,v,"b-",label="v(t)")
    plt.xlabel("time (s)",fontsize=20)
    plt.ylabel("y(t),v(t)",fontsize=20)
    plt.xlim(0,20.)
    plt.ylim(-100.,600.)
    plt.grid()
    plt.legend(loc="upper left", bbox_to_anchor=(1,1), fontsize=20)
    plt.show()
```

```python
def EulerCromerFreeFall():
    tmax=10
    dt=0.1
    a=-9.81
    nsteps=int(tmax/dt)
    t = np.zeros(nsteps)
    v = np.zeros(nsteps)
    y = np.zeros(nsteps)
    t[0]=0
    v[0]=0
    for i in range(len(t)-1):
        t[i+1]=t[i]+dt
        v[i+1]=v[i]+dt*a
        y[i+1]=y[i]+dt*v[i+1]

    return t,v,y
```

```python
tec,vec,yec=EulerCromerFreeFall()
```

```python
def PPderiv(y,t):
    alpha=0.7
    gamma=0.5
    beta=0.007
    eps=0.1
    Zi=y[0]
    Li=y[1]
    dzdt=alpha*Zi-beta*Zi*Li
    dldt=eps*beta*Zi*Li-gamma*Li
    return np.array([dzdt,dldt])
```

```python
t=np.arange(0.,50.+0.1,0.1)
z0=200
l0=50
initsLZ=np.array([z0,l0])
result=odeint(PPderiv,np.array([z0,l0]),t)
z=result[:,0]
l=result[:,1]
plt.plot(t,z,label='zebra pop')
plt.plot(t,l,label='lion pop')
plt.xlabel('time(years)')
plt.ylabel('population')
plt.legend()
plt.show()
```
